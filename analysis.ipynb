```python
import pandas as pd 
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\System -02\Desktop\Project\lpg_messy_dataset.csv')
print(f"Structure of table  : {df.shape}")
print(f"Columns name :    {df.columns}")
print(f"Total rows included duplicates : {df.isnull().sum().sort_values(ascending=False).head(3)}")
```

```python
# Droped duplicates
df.drop_duplicates(inplace = True)
print(f"The total num of rows and columns :{df.shape}")

#Handled unconsinent value in state column
df['State'] = df['State'].str.replace({"up":"Uttar Pradesh",
                                       "Uttar Pardesh":"Uttar Pradesh", })

# Handle mising value which is almost same so we used ffill .which basically pick the last value means above value of null
cols = ['Demand_MT','Price_INR']
df[cols] = df[cols].ffill() 
print(f"Preview of nulls :{df.isnull().sum().sort_values(ascending=False).head(3)}")

#Converting data fromats and making new columns of dates for better visualization
df['Date'] = pd.to_datetime(df['Date'])
df['Time'] = df['Date'].dt.time

#Handeling price null values,outliers and changing the data type as well
df['Price_INR'] = pd.to_numeric(df['Price_INR'],errors='coerce')
df.loc[df['Price_INR'] == 9999,'Price_INR'] = np.nan
df['Price_INR'] =df['Price_INR'].ffill()
df['Price_INR'] =df['Price_INR'].fillna(df['Price_INR'].median())

# Seeing unique price
print(df['Price_INR'].unique())
```
